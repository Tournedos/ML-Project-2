```python
def load_lifespan(pathin):
    subfolders = ['control','companyDrug']
    feat_dict = {}
    for subf in subfolders:
        subfp = os.path.join(pathin, subf)
        filenms = os.listdir(subfp)
        for cont, name in enumerate(filenms):
            filepath = os.path.join(subfp,name)
            print(filepath)
            try:
                # Try reading as a CSV
                data_raw = pd.read_csv(filepath, sep=',')
                print("File loaded as CSV.")
            except Exception as e:
                continue
                #print(f"Failed to load as CSV. Error: {e}")
                #try:
                #    # If CSV fails, try reading as Excel
                #    data_raw = pd.read_excel(filepath)
                #    print("File loaded as Excel.")
                #except Exception as e:
                #    print(f"Failed to load as Excel. Error: {e}")
                #    raise ValueError("File could not be loaded. Please check the format.")
            #print(np.array(data_raw.head()).T)
            sample_name = 'worm_' + str(cont) + '_' + subf
            data_n = data_raw.apply(pd.to_numeric)
            feat_dict.update({sample_name: np.array(data_n).T})
    return feat_dict
```

```python
# Function to count successive missing values
def count_successive_missing(array):
    missing_mask = np.isnan(array)
    successive_counts = np.diff(np.concatenate(([False], missing_mask.any(axis=0), [False])).astype(int))
    start_indices = np.where(successive_counts == 1)[0]
    end_indices = np.where(successive_counts == -1)[0]
    lengths = end_indices - start_indices
    return list(zip(start_indices, end_indices, lengths))

# Iterate over the dictionary
for name, lifespan_array in fdict.items():
    print(f"Processing {name}")
    lifespan_arrayxy = lifespan_array[2:4,:]
    missing_sequences = count_successive_missing(lifespan_array)
    for start, end, length in missing_sequences:
        print(f"  Missing sequence starts at column {start}, ends at column {end - 1}, length: {length}")
```

```python
# Function to cut array based on missing values in rows 2:4 (zero-indexed)
def cut_array(array, rows_to_check):
    # Check for NaNs in the specified rows (2:4 corresponds to rows 2, 3, 4)
    missing_mask = ~np.isnan(array[rows_to_check, :]).any(axis=0)  # True for valid columns
    return array[:, missing_mask]  # Retain all rows, only valid columns

# Rows to check for missing values (2:4 in zero-based indexing)
rows_to_check = slice(2, 4)  # Rows 2, 3, 4

# Process all arrays in the dictionary
cut_lifespan_dict = {name: cut_array(array, rows_to_check) for name, array in fdict.items()}
```
